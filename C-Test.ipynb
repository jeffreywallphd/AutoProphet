```python
# Initialize the variables
CIKNumber = '0000027419'
filingType = "10-K"
start_date = datetime(2021, 1, 1)
end_date = datetime(2021, 12, 30)

# Create widgets for each variable
cik_number_widget = widgets.Text(value=CIKNumber, description='CIK Number:') 

filing_type_widget = widgets.Dropdown(
    options=['10-K', '10-Q'],
    value=filingType,
    description='Filing Type:'
)
start_date_widget = widgets.DatePicker(value=start_date, description='Start Date:')
end_date_widget = widgets.DatePicker(value=end_date, description='End Date:')

# Create a button widget
update_button = widgets.Button(description="Update Variables")

# Display the widgets and button
display(cik_number_widget)
display(filing_type_widget)
display(start_date_widget)
display(end_date_widget)
display(update_button)

output = widgets.Output()

# Function to update the variables
@output.capture()
def update_variables(change):
    global CIKNumber, filingType, start_date, end_date
    new_CIKNumber = cik_number_widget.value
    if len(new_CIKNumber) != 10:
        with output:
            clear_output(wait=True)  # Clear the previous output and wait for new output
            print("CIK Number must have exactly 10 characters.")
    else:
        global CIKNumber, filingType, start_date, end_date
        CIKNumber = new_CIKNumber
        filingType = filing_type_widget.value
        start_date = start_date_widget.value
        end_date = end_date_widget.value
        with output:
            clear_output(wait=True)  # Clear the previous output and wait for new output
            print("Variables updated successfully.")
            time.sleep(2)  # Wait for 2 seconds
            clear_output()  # Clear the success message

# Attach the update function to the button click event
update_button.on_click(update_variables)
display(output)
```

```python
class SecFilingDownloader:
    def __init__(self, CIKNumber, filingType, start_date, end_date):
        self.baseURL = "https://www.sec.gov"
        self.CIKNumber = CIKNumber
        self.filingType = filingType
        self.start_date = start_date
        self.end_date = end_date
        self.header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/105.0.0.0 Safari/537.36'}
        self.apiData = None
        self.data = {"statement": []}
    
    def download_api_data(self):
        try:
            apiResponse = requests.get(f"https://data.sec.gov/api/xbrl/companyfacts/CIK{self.CIKNumber}.json", headers=self.header)
            apiResponse.raise_for_status()
            self.apiData = json.loads(apiResponse.content)
        except requests.exceptions.RequestException as e:
            print(f"API Request Error: {e}")
        except json.JSONDecodeError as e:
            print(f"JSON Decode Error: {e}")
    
    def download_company_filing(self):
        if self.apiData:
            filingParameters = {
                'action': 'getcompany',
                'CIK': self.CIKNumber,
                'type': self.filingType,
                'dateb': self.end_date,
                'datea': self.start_date,
                'owner': 'exclude',
                'output': '',
                'count': '100'}
            try:
                companyResponse = requests.get(url=r"https://www.sec.gov/cgi-bin/browse-edgar", params=filingParameters,
                                            headers=self.header)
                companyResponse.raise_for_status()
                companyParser = BeautifulSoup(companyResponse.content, 'html.parser')
                documentLink = companyParser.find("a", {"id": "documentsbutton"})
                if documentLink:
                    documentURL = self.baseURL + documentLink["href"]
                    print(f"Document Link: {documentURL}")
                    self.download_document(documentURL)
                else:
                    print("Document link not found.")
            except requests.exceptions.RequestException as e:
                print(f"Company or Calculation Report Request Error: {e}")
            except Exception as e:
                print(f"Error: {e}")
        else:
            print("API data not available.")
    
    def download_document(self, documentURL):
        try:
            documentResponse = requests.get(documentURL, headers=self.header)
            documentResponse.raise_for_status()
            documentParser = BeautifulSoup(documentResponse.content, 'html.parser')
            xmlTable = documentParser.find("table", {"summary": "Data Files"})
            xmlTableData = xmlTable.find_all("td", {"scope": "row"})
            dataFilesTable = documentParser.find("table", summary="Data Files")
            if dataFilesTable:
                rows = dataFilesTable.find_all("tr")
                for row in rows:
                    columns = row.find_all("td")
                    if len(columns) >= 3 and "EXTRACTED XBRL INSTANCE DOCUMENT" in columns[1].text:
                        xbrlInstanceLink = self.baseURL + columns[2].find("a")["href"]
                        print(f"XBRL Instance Link: {xbrlInstanceLink}")
                        self.download_xbrl_instance(xbrlInstanceLink)
                        calculationLink = self.find_calculation_link(documentParser)
                        break
                else:
                    print("XBRL Instance data not found.")
            else:
                print("XBRL Instance data not found.")
        except requests.exceptions.RequestException as e:
            print(f"Document Request Error: {e}")
        except Exception as e:
            print(f"Error: {e}")
    
    def find_calculation_link(self, documentParser):
        try:
            xmlTableData = documentParser.find_all("td", {"scope": "row"})
            if xmlTableData:
                for i in range(len(xmlTableData) - 1):
                    if xmlTableData[i].getText() == "XBRL TAXONOMY EXTENSION CALCULATION LINKBASE DOCUMENT":
                        xbrlCalculationData = xmlTableData[i + 1].find("a")
                        if xbrlCalculationData:
                            calculationLink = self.baseURL + xbrlCalculationData["href"]
                            print(f"XBRL Calculation Linkbase: {calculationLink}")
                            return calculationLink
                        else:
                            print("XBRL Calculation data not found.")
                            return None
                else:
                    print("XBRL Calculation data not found.")
                    return None
            else:
                print("XBRL Calculation data not found.")
                return None
        except requests.exceptions.RequestException as e:
            print(f"XBRL Calculation Request Error: {e}")
        except Exception as e:
            print(f"Error: {e}")
            return None
    
    def download_xbrl_instance(self, xbrlInstanceLink):
        try:
            xbrlInstanceResponse = requests.get(xbrlInstanceLink, headers=self.header)
            xbrlInstanceResponse.raise_for_status()
            xbrlInstanceParser = BeautifulSoup(xbrlInstanceResponse.content, 'html.parser')
            # Add logic to process the XBRL instance document here
        except requests.exceptions.RequestException as e:
            print(f"XBRL Instance Request Error: {e}")
        except Exception as e:
            print(f"Error: {e}")
    
    def download_schema(self, schemaLink):
        try:
            schemaReportResponse = requests.get(schemaLink, headers=self.header)
            schemaReportResponse.raise_for_status()
            schemaReportParser = BeautifulSoup(schemaReportResponse.content, features="xml")
            financialStatements = schemaReportParser.find_all("link:calculationLink", {"xlink:type": "extended"})
            # Add logic to process the schema document here
        except requests.exceptions.RequestException as e:
            print(f"Schema Request Error: {e}")
        except Exception as e:
            print(f"Error: {e}")
```

```python
# Sample data (replace with your actual data)
data = {
    "Concept": [
        "AccountsAndOtherReceivablesNetCurrent",
        "AccountsAndOtherReceivablesNetCurrent",
        # Add more data
    ],
    "Report Type": ["10-K", "10-K"],
    "Report Number": ["0000027419-22-000007", "0000027419-22-000007"],
    "Fiscal Year": [2021, 2021],
    "Period Ending": ["2021-01-30", "2022-01-29"],
    "Value": ["$631,000,000.00", "$835,000,000.00"],
}

df = pd.DataFrame(data, columns=["Concept", "Report Type", "Report Number", "Fiscal Year", "Period Ending", "Value"])

# Define a function to construct financial statements
def construct_financial_statements(df, categories, fiscal_year, period_ending):
    statements = {}
    
    for category in categories:
        filtered_df = df[df['Concept'].str.contains(category, case=False, na=False)]
        filtered_df = filtered_df[(filtered_df['Fiscal Year'] == fiscal_year) & (filtered_df['Period Ending'] == period_ending)]
        
        if not filtered_df.empty:
            # Process the filtered DataFrame, e.g., sum the values
            total_value = filtered_df['Value'].str.replace('[\$,]', '', regex=True).astype(float).sum()
            statements[category] = total_value
    
    return statements

# Define your financial statement categories
financial_categories = ["Income Statement", "Balance Sheet", "Cash Flow Statement"]

# Create interactive widgets
fiscal_year_dropdown = widgets.Dropdown(
    options=df['Fiscal Year'].unique(),
    description='Fiscal Year:',
    layout=widgets.Layout(width='200px')
)

period_ending_dropdown = widgets.Dropdown(
    options=df['Period Ending'].unique(),
    description='Period Ending:',
    layout=widgets.Layout(width='200px')
)

update_button = widgets.Button(description='Generate Financial Statements')

# Define a function to handle button click event
def on_button_click(b):
    fiscal_year = fiscal_year_dropdown.value
    period_ending = period_ending_dropdown.value
    statements = construct_financial_statements(df, financial_categories, fiscal_year, period_ending)
    clear_output(wait=True)
    
    # Create a formatted string to display the statements
    statement_str = ""
    for category, total_value in statements.items():
        statement_str += f"{category}: ${total_value:,.2f}\n"
    
    display(Markdown(statement_str))

# Attach the click event handler to the button
update_button.on_click(on_button_click)

# Display widgets and initial DataFrame
widget_box = widgets.HBox([fiscal_year_dropdown, period_ending_dropdown, update_button])
display(widget_box)
```
